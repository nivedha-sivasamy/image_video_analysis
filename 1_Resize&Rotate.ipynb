#**Program to implement Resize, Rotation of binary, gray-scale and color images using various methods**

```python
import cv2
from google.colab.patches import cv2_imshow
import matplotlib.pyplot as plt
import imutils
```

```python
class Processing:

  def __init__(self,path):
    self.path=path
    self.image=cv2.imread(self.path,1)
    self.height=self.image.shape[0]
    self.width=self.image.shape[1]

  def image_choice(self):
    c_img=int(input("The choice of image :\n1.Colour\n2.Gray\n3.Binary\n: "))
    if c_img not in [1,2,3]:
      print('Enter valid choice')
      self.image_choice()
    elif c_img==1:
      img1=cv2.imread(self.path,1)
      return img1
    elif c_img==2:
      gray_img=cv2.imread(self.path,0)
      return gray_img
    elif c_img==3:
      gray_img=cv2.imread(self.path,0)
      _,bin_img=cv2.threshold(gray_img,127,255,cv2.THRESH_BINARY)
      return bin_img

  def aspect_resize(self,image):
    scale=int(input('\nEnter the percentage to scale the image : '))
    width_ratio=int(self.width*scale/100)
    height_ratio=int(self.height*scale/100)
    dsize=(width_ratio,height_ratio)
    r_img=cv2.resize(image,dsize)
    return scale,r_img

  def c_resize(self,image):
    width=int(input('\nEnter the width for the image : '))
    height=int(input('Enter the height for the image : '))
    dsize=(width,height)
    r_img=cv2.resize(image,dsize)
    return width,height,r_img

  def rotate(self,image):
    print("\nRotation")
    angle=int(input('\nEnter the angle for rotation : '))
    direction=int(input('Clockwise (1) or anti-clockwise (2): '))
    if direction==1:
      ro_img=imutils.rotate_bound(image,-angle)
    elif direction==2:
      ro_img=imutils.rotate_bound(image,angle)
    return angle,direction,ro_img
```

```python
def main():
  print('Resizing and rotation operations')
  p=Processing('/content/drive/MyDrive/marvel/train/spider-man/pic_236.jpg')
  while True:
    print("\nOperations:\n1.Rotation\n2.Percentage_Resizing\n3.Custom_Resizing\n4.Exit\n")
    c=int(input("Your choice : "))
    if c==1:
      r_img=p.image_choice()
      a,d,r_img1=p.rotate(r_img)
      print(f'\nOriginal Image Height {p.height} Width {p.width}\n')
      cv2_imshow(r_img)
      if d==1:
        d1='Clockwise'
      else:
        d1='Anti-clockwise'
      print(f"\nImage after rotation : Angle {a}, Direction {d1}\n")
      cv2_imshow(r_img1)
    elif c==2:
      print('\nPercentage Resizing')
      re_img=p.image_choice()
      s,re_img1=p.aspect_resize(re_img)
      print(f'\nOriginal Image Width {p.width}, Height {p.height}\n')
      cv2_imshow(re_img)
      height1=re_img1.shape[0]
      width1=re_img1.shape[1]
      print(f"\nImage after resizing Percent {s}% Width {width1}, Height {height1}\n")
      cv2_imshow(re_img1)
    elif c==3:
      print('\nCustom Resizing')
      cre_img=p.image_choice()
      w,h,cre_img1=p.c_resize(cre_img)
      print(f'\nOriginal Image Width {p.width}, Height {p.height}\n')
      cv2_imshow(cre_img)
      print(f"\nImage after resizing Width {w}, Height {h}\n")
      cv2_imshow(cre_img1)
    elif c==4:
      print("\nEND\n")
      break
    else:
      print("\nEnter valid choice!!")
```

```python
main()
```
