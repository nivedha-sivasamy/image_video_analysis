```python
import cv2
import numpy as np
from google.colab.patches import cv2_imshow
import matplotlib.pyplot as plt
```

```python
o_img=cv2.imread('/content/drive/MyDrive/marvel/train/ironman/pic_364.jpg')
n_img=cv2.imread('/content/drive/MyDrive/marvel/train/ironman/pic_364_noise.jpg')
```

```python
# Image Denoising
def denoise(img):
  print('Original Image')
  cv2_imshow(img)
  img1=cv2.fastNlMeansDenoisingColored(img,None,19,17)
  print('\nImage after Denoising')
  cv2_imshow(img1)
```

```python
def smoothening(img):
  print('Original Image')
  cv2_imshow(img)
  e=cv2.edgePreservingFilter(img,flags=1,sigma_s=155,sigma_r=0.45)
  print('\nImage after smoothening')
  cv2_imshow(e)
```

```python
def blur(img):
  print('Original Image')
  cv2_imshow(img)
  blur=cv2.medianBlur(img,5)
  print('\nImage after blurring')
  cv2_imshow(blur)
```

```python
# Edge detection
def edge(img):
  cv2_imshow(img)
  e_filter = np.array([[0, -1, 0], [-1, 4, -1], [0, -1, 0]])
  e_img = cv2.filter2D(img, -1, e_filter)
  print('\nEdge Detection')
  cv2_imshow(e_img)
```

```python
def sharp(img):
  blur=cv2.medianBlur(img,5)
  cv2_imshow(blur)
  s_filter = np.array([[0, -1, 0], [-1, 5, -1], [0, -1, 0]])
  s_img = cv2.filter2D(blur, -1, s_filter)
  print('\nSharpen image')
  cv2_imshow(s_img)
```

```python
def main(o_img,n_img):
  print('Image Enhancement techniques')
  while True:
    print('Options:\n1.Denoising\n2.Smoothening\n3.Blurring\n4.Edge Detection\n5.Sharpen\n6.Exit')
    c=int(input('Your choice : '))
    if c==1:
      print('Denoising\n')
      denoise(n_img)
    elif c==2:
      print('Smoothening\n')
      smoothening(o_img)
    elif c==3:
      print('Blurring\n')
      blur(o_img)
    elif c==4:
      print('Edge detection\n')
      edge(o_img)
    elif c==5:
      sharp(o_img)
    elif c==6:
      print('End')
      break
    else:
      print('Invalid Choice!')
      continue
```

```python
main(o_img,n_img)
```
