#**Program to implement Non-Linear Spatial Filter using Built in and user defined functions.**

```python
import cv2
import numpy as np
from google.colab.patches import cv2_imshow
from skimage.util import random_noise
import matplotlib.pyplot as plt
```

```python
def compare(img,img1,filter):
  fig,ax=plt.subplots(1,2,figsize=(10,10))
  ax[0].set_title('Original Image')
  ax[0].imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
  ax[1].set_title(f'Image after applying {filter}')
  ax[1].imshow(cv2.cvtColor(img1, cv2.COLOR_BGR2RGB))
```

```python
def median_blur(img):
  median = cv2.medianBlur(img,5)
  return img,median
```

```python
def gaussian_blur(img):
  blur = cv2.GaussianBlur(img,(7,7),0)
  return blur
```

```python
def min_filter(img):
  size = (3, 3)
  shape = cv2.MORPH_RECT
  kernel = cv2.getStructuringElement(shape, size)
  min_image = cv2.erode(img, kernel)
  return min_image
```

```python
def max_filter(img):
  size = (3, 3)
  shape = cv2.MORPH_RECT
  kernel = cv2.getStructuringElement(shape, size)
  max_image = cv2.dilate(img, kernel)
  return max_image
```

```python
def bilateral(img):
  bi=cv2.bilateralFilter(img,15,80,80)
  return bi
```

```python
img=cv2.imread('/content/drive/MyDrive/marvel/train/black widow/pic_004.jpg')
def main(img):
  print('Different types of Non-linear spatial filters')
  n,mb=median_blur(img)
  compare(n,mb,'Median Blur')
  gb=gaussian_blur(img)
  compare(img,gb,'Gaussian Blur')
  mf=min_filter(img)
  compare(img,mf,'Minimum Filter')
  maf=max_filter(img)
  compare(img,maf,'Maximum Filter')
  bi=bilateral(img)
  compare(img,bi,'Bilateral Filter')
```

```python
main(img)
```
