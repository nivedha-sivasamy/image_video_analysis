```python
import cv2
import os
from google.colab.patches import cv2_imshow
import datetime
import glob
import math
import random
```

```python
def frames(video):
  fps = math.ceil(video.get(cv2.CAP_PROP_FPS))
  print('Frames per second:',fps)
  total = int(video.get(cv2.CAP_PROP_FRAME_COUNT))
  print("Total number of frames:",total)
  seconds=(total/fps)
  video_time = datetime.timedelta(seconds=seconds)
  print("Total video length in seconds:",seconds,"sec")
  print('Video Duration',video_time)
  currentframe = 0
  lst=[]
  while(True):
      ret,frame = video.read()
      if ret:
        cv2.imwrite('/content/drive/MyDrive/marvel/train/images/image'+
                    str(currentframe)+'.jpg',frame)
        currentframe += 1
        lst.append(frame)
        # print('\nFrame ',currentframe,'\n')
        # cv2_imshow(frame)
      else:
        break

def video_from_frames(img,size):
  original_video = cv2.VideoWriter(
      '/content/drive/MyDrive/marvel/train/images/videos/output_video.avi',
      cv2.VideoWriter_fourcc(*'DIVX'), 10, size)
  for i in range(len(img)):
    original_video.write(img[i])
  original_video.release()

def slow_video(img,size):
  slow_video =cv2.VideoWriter(
      "/content/drive/MyDrive/marvel/train/images/videos/slow_video.mp4",
      cv2.VideoWriter_fourcc(*"DIVX"), 2, size)
  for i in range(len(img)):
    slow_video.write(img[i])

def fast_video(img_array,size):
  fast_video =cv2.VideoWriter(
      "/content/drive/MyDrive/marvel/train/images/videos/fast_video.avi",
      cv2.VideoWriter_fourcc(*"DIVX"), 18, size)
  for i in range(len(img_array)):
    fast_video.write(img_array[i])

def reverse(size):
  video = cv2.VideoCapture(
      "/content/drive/MyDrive/marvel/train/images/videos/loki-tesseract.gif")
  frame_list = []
  while(True):
      ret,frame = video.read()
      if ret:
        frame_list.append(frame)
      else:
        break
  frame_list.pop()
  frame_list.reverse()
  reverse_original_video = cv2.VideoWriter(
      '/content/drive/MyDrive/marvel/train/images/videos/reversed_video.mp4',
      cv2.VideoWriter_fourcc(*'DIVX'), 2, size)
  for i in range(len(frame_list)):
    reverse_original_video.write(frame_list[i])
  reverse_original_video.release()


def mixed_video(size):
  video = cv2.VideoCapture(
      "/content/drive/MyDrive/marvel/train/images/videos/loki-tesseract.gif")
  img = []
  while(True):
      ret,frame = video.read()
      if ret:
        img.append(frame)
      else:
        break
  mixed_video = cv2.VideoWriter(
      '/content/drive/MyDrive/marvel/train/images/videos/mixed_video.mp4',
      cv2.VideoWriter_fourcc(*'DIVX'), 2, size)
  random.shuffle(img)
  for i in range(len(img)):
    mixed_video.write(img[i])
  mixed_video.release()

def change_colour(size):
  video = cv2.VideoCapture(
      "/content/drive/MyDrive/marvel/train/images/videos/loki-tesseract.gif")
  img = []
  currentframe=0
  while(True):
      ret,frame = video.read()
      if ret:
        color = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
        cv2.imwrite('/content/drive/MyDrive/marvel/train/images/g_images/image'
        +str(currentframe)+'.jpg',color)
        img.append(color)
        currentframe+=1
        # print(f'\nFrame {currentframe}\n')
        # cv2_imshow(color)
      else:
        break
  c_img_array = []
  for filename in glob.glob(
      '/content/drive/MyDrive/marvel/train/images/g_images/*.jpg'):
      c_img = cv2.imread(filename)
      height, width, layers = c_img.shape
      size = (width,height)
      c_img_array.append(c_img)

  c_video = cv2.VideoWriter(
      '/content/drive/MyDrive/marvel/train/images/videos/gray_video.avi',
      cv2.VideoWriter_fourcc(*'DIVX'), 10, size)
  for i in range(len(c_img_array)):
    c_video.write(c_img_array[i])

  c_video.release()

def rotate(size):
  video = cv2.VideoCapture(
      "/content/drive/MyDrive/marvel/train/images/videos/loki-tesseract.gif")
  img = []
  currentframe=0
  while(True):
      ret,frame = video.read()
      if ret:
        r = cv2.rotate(frame, cv2.ROTATE_180)
        cv2.imwrite('/content/drive/MyDrive/marvel/train/images/r_images/image'
        +str(currentframe)+'.jpg',r)
        img.append(r)
        currentframe+=1
        # print(f'\nFrame {currentframe}\n')
        # cv2_imshow(r)
      else:
        break
  r_img_array = []
  for filename in glob.glob(
      '/content/drive/MyDrive/marvel/train/images/r_images/*.jpg'):
      r_img = cv2.imread(filename)
      height, width, layers = r_img.shape
      size = (width,height)
      r_img_array.append(r_img)

  r_video = cv2.VideoWriter(
      '/content/drive/MyDrive/marvel/train/images/videos/rotate_video.avi',
      cv2.VideoWriter_fourcc(*'DIVX'), 10, size)
  for i in range(len(r_img_array)):
    r_video.write(r_img_array[i])

  r_video.release()
```

```python
def main():
  print("**Video Analysis**")

  print("Extracting Frames from Videos...")
  vid = cv2.VideoCapture(
      "/content/drive/MyDrive/marvel/train/images/videos/loki-tesseract.gif")
  frames(vid)
  print("\nFrames Extracted")
  img_array = []
  for filename in glob.glob('/content/drive/MyDrive/marvel/train/images/*.jpg'):
    img = cv2.imread(filename)
    height, width, layers = img.shape
    size = (width,height)
    img_array.append(img)
  while True:
    print("\n1.Creating Video from frames\n2.Slow Motion Video")
    print("3.Fast Motion Video\n4.Reversed Video\n5.Mixed Video")
    print("6.Color change video\n7.Rotate Video\n8.Exit")
    choice = int(input("Enter your choice: "))
    if choice ==1:
      print("\n\nCoverting Frames to Videos..")
      video_from_frames(img_array,size)
    elif choice ==2:
      slow_video(img_array,size)
      print("\nSlow Motion Video Created.")
    elif choice ==3:
      fast_video(img_array, size)
      print("\nFast Motion Video Created.")
    elif choice ==4:
      reverse(size)
      print("\nReverse Motion Video Created.")
    elif choice==5:
      mixed_video(size)
      print('\nMixed video created')
    elif choice==6:
      change_colour(size)
      print('\nVideo created\n')
    elif choice==7:
      rotate(size)
      print('\nRotated Video created\n')
    elif choice==8:
      print("Done")
      break
    else:
      print("Invalid Choice:")
      continue

main()

```
